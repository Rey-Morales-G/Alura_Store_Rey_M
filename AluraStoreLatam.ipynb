```python
# sales by product
ventas_1 = tienda1.groupby("Producto").size().sort_values(ascending=False)
ventas_2 = tienda2.groupby("Producto").size().sort_values(ascending=False)
ventas_3 = tienda3.groupby("Producto").size().sort_values(ascending=False)
ventas_4 = tienda4.groupby("Producto").size().sort_values(ascending=False)

# Best y worst (Top 1 y Bottom 1)
best_1 = ventas_1.head(1)
best_2 = ventas_2.head(1)
best_3 = ventas_3.head(1)
best_4 = ventas_4.head(1)

worst_1 = ventas_1.tail(1)
worst_2 = ventas_2.tail(1)
worst_3 = ventas_3.tail(1)
worst_4 = ventas_4.tail(1)

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

best_values = [
    int(best_1.iloc[0]),
    int(best_2.iloc[0]),
    int(best_3.iloc[0]),
    int(best_4.iloc[0]),
]

worst_values = [
    int(worst_1.iloc[0]),
    int(worst_2.iloc[0]),
    int(worst_3.iloc[0]),
    int(worst_4.iloc[0]),
]

best_names = [
    best_1.index[0],
    best_2.index[0],
    best_3.index[0],
    best_4.index[0],
]

worst_names = [
    worst_1.index[0],
    worst_2.index[0],
    worst_3.index[0],
    worst_4.index[0],
]

import numpy as np

x = np.arange(len(tiendas))
width = 0.35

plt.figure(figsize=(10,6))
plt.bar(x - width/2, best_values, width, label="Best seller")
plt.bar(x + width/2, worst_values, width, label="Worst seller")

plt.xticks(x, tiendas)
plt.ylabel("Cantidad de ventas")
plt.title("Best vs Worst seller por tienda")
plt.legend()

# labels
for i in range(len(tiendas)):
    # Best
    plt.text(x[i] - width/2, best_values[i], str(best_values[i]),
             ha="center", va="bottom", fontsize=8)
    # Worst
    plt.text(x[i] + width/2, worst_values[i], str(worst_values[i]),
             ha="center", va="bottom", fontsize=8)

plt.tight_layout()
plt.show()

# get info on what is the product name
for i, tienda in enumerate(tiendas):
    print(f"{tienda} - Best seller: {best_names[i]} ({best_values[i]} ventas)")
    print(f"{tienda} - Worst seller: {worst_names[i]} ({worst_values[i]} ventas)")
    print("-" * 60)
```
