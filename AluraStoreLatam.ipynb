# 2. Ventas por categoría

## Grouping and counting the categories ##

```python
categorias_1 = tienda1.groupby("Categoría del Producto").size().sort_values(ascending=False)
categorias_2 = tienda2.groupby("Categoría del Producto").size().sort_values(ascending=False)
categorias_3 = tienda3.groupby("Categoría del Producto").size().sort_values(ascending=False)
categorias_4 = tienda4.groupby("Categoría del Producto").size().sort_values(ascending=False)

categorias_1, categorias_2, categorias_3, categorias_4
```

## Results compared side by side for easy readibility ##

# 3. Calificación promedio de la tienda


## calculating the client satisfaction score by store ##

```python
# Average client rating for each store

promedio_1 = tienda1["Calificación"].mean()
promedio_2 = tienda2["Calificación"].mean()
promedio_3 = tienda3["Calificación"].mean()
promedio_4 = tienda4["Calificación"].mean()

print("Promedio de Calificación - Tienda 1:", round(promedio_1, 2))
print("Promedio de Calificación - Tienda 2:", round(promedio_2, 2))
print("Promedio de Calificación - Tienda 3:", round(promedio_3, 2))
print("Promedio de Calificación - Tienda 4:", round(promedio_4, 2))
```

### table for visualization purposes ###

```python
import pandas as pd

promedios = pd.DataFrame({
    "Promedio Calificación": [
        promedio_1, promedio_2, promedio_3, promedio_4
    ]
}, index=["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"])

promedios.round(2)
```

# 4. Productos más y menos vendidos

## Best and Worst Sellers ##

Let's group products, count their sale occurence and sort it

```python
# Count product sales per store

ventas_1 = tienda1.groupby("Producto").size().sort_values(ascending=False)
ventas_2 = tienda2.groupby("Producto").size().sort_values(ascending=False)
ventas_3 = tienda3.groupby("Producto").size().sort_values(ascending=False)
ventas_4 = tienda4.groupby("Producto").size().sort_values(ascending=False)
```

get the top and worst sellers

```python
print("🟩 TOP 5 Best Sellers – Tienda 1:")
print(ventas_1.head(5), "\n")

print("🟥 WORST 5 Sellers – Tienda 1:")
print(ventas_1.tail(5), "\n")

print("🟩 TOP 5 Best Sellers – Tienda 2:")
print(ventas_2.head(5), "\n")

print("🟥 WORST 5 Sellers – Tienda 2:")
print(ventas_2.tail(5), "\n")

print("🟩 TOP 5 Best Sellers – Tienda 3:")
print(ventas_3.head(5), "\n")

print("🟥 WORST 5 Sellers – Tienda 3:")
print(ventas_3.tail(5), "\n")

print("🟩 TOP 5 Best Sellers – Tienda 4:")
print(ventas_4.head(5), "\n")

print("🟥 WORST 5 Sellers – Tienda 4:")
print(ventas_4.tail(5), "\n")
```

## investigating how to better make this list ##

```python
ventas_1 = tienda1.groupby("Producto").size().sort_values(ascending=False)
ventas_2 = tienda2.groupby("Producto").size().sort_values(ascending=False)
ventas_3 = tienda3.groupby("Producto").size().sort_values(ascending=False)
ventas_4 = tienda4.groupby("Producto").size().sort_values(ascending=False)

best_1 = ventas_1.head(1)
best_2 = ventas_2.head(1)
best_3 = ventas_3.head(1)
best_4 = ventas_4.head(1)

worst_1 = ventas_1.tail(1)
worst_2 = ventas_2.tail(1)
worst_3 = ventas_3.tail(1)
worst_4 = ventas_4.tail(1)

import pandas as pd

best_worst_table = pd.DataFrame({
    "Best Seller": [
        best_1.index[0],
        best_2.index[0],
        best_3.index[0],
        best_4.index[0],
    ],
    "Best Seller – Sales": [
        int(best_1.iloc[0]),
        int(best_2.iloc[0]),
        int(best_3.iloc[0]),
        int(best_4.iloc[0]),
    ],
    "Worst Seller": [
        worst_1.index[0],
        worst_2.index[0],
        worst_3.index[0],
        worst_4.index[0],
    ],
    "Worst Seller – Sales": [
        int(worst_1.iloc[0]),
        int(worst_2.iloc[0]),
        int(worst_3.iloc[0]),
        int(worst_4.iloc[0]),
    ],
}, index=["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"])

best_worst_table
```
