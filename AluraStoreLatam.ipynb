```python
import numpy as np
import matplotlib.pyplot as plt

# Data already calculated in your notebook:
# revenue, ratings, shipping, best_values, worst_values

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Convert to numpy arrays for normalization
revenue = np.array(revenue)
ratings = np.array(ratings)
shipping = np.array(shipping)
best = np.array(best_values)
worst = np.array(worst_values)

# Normalize all metrics to 0–1 range
def normalize(x, invert=False):
    if invert:
        x = max(x) - x + min(x)
    return (x - min(x)) / (max(x) - min(x))

revenue_n = normalize(revenue)
ratings_n = normalize(ratings)
shipping_n = normalize(shipping, invert=True)   # lower shipping = better
best_n = normalize(best)
worst_n = normalize(worst, invert=True)        # lower worst seller = worse

# Put them together
metrics = np.vstack([revenue_n, ratings_n, shipping_n, best_n, worst_n])

labels = ["Facturación", "Rating", "Envío (inv.)", "Best Seller", "Worst (inv.)"]
num_vars = len(labels)

# Angles for radar
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

plt.figure(figsize=(9, 9))

# Plot each store
for i in range(len(tiendas)):
    values = metrics[:, i].tolist()
    values += values[:1]  # close the loop
    plt.polar(angles, values, marker='o', linewidth=2, label=tiendas[i])

# Title & Legend
plt.title("Radar Chart – Comparación Total de KPIs por Tienda", size=16)
plt.xticks(angles[:-1], labels)
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.show()


```

###  Recommendation: Divestment of Tienda 4

After analyzing sales performance, customer satisfaction, product mix, and shipping efficiency across the four stores, **Tienda 4 appears as the weakest performer** and the primary candidate for divestment.

---

## 1. Total Sales

Tienda 4 generated the **lowest overall revenue** with **1,038,375,700.00**, falling behind:

- Tienda 1: 1,150,880,400.00  
- Tienda 2: 1,116,343,500.00  
- Tienda 3: 1,098,019,600.00  

This consistent gap indicates **lower demand** and **reduced revenue potential** compared with the rest of the network.

---

## 2. Best and Worst Selling Products

Although Tienda 4 has a solid best seller, its product portfolio shows **limited depth**, with the worst-selling product reaching only **33 units**, similar to or worse than other stores.

This suggests:

- slow product turnover  
- uneven product-market fit  
- a weaker overall assortment performance  

The best sellers in other stores show stronger demand levels, widening the competitive gap.

---

## 3. Product Categories Performance

The analysis of product categories shows that Tienda 4:

- sells fewer units across its leading categories  
- depends heavily on a small number of product types  
- lacks representation in some of the strongest categories from other stores  

This indicates that **customer demand in Tienda 4's area is either weaker or misaligned with the offered assortment**, reducing growth potential.

---

## 4. Customer Satisfaction

Tienda 4 has an average rating of **3.99**, which is:

- lower than Tienda 2 (4.04)  
- lower than Tienda 3 (4.05)  
- slightly higher than Tienda 1 (3.98)

Although not the worst, the rating **does not compensate for the store’s lower sales performance**, showing average customer satisfaction but poor market traction.

---

## 5. Shipping Efficiency

Tienda 4 has the **lowest shipping cost**, at **23,459.46**, meaning it has a **cost advantage**.  
However, despite this advantage:

- revenue remains the lowest in the network  
- categories do not perform strongly  
- ratings are not significantly higher  

This indicates that **lower logistics cost is not enough to offset weak demand**.

---

## Final Conclusion

Considering:

- **Lowest total sales** in the network  
- **Weaker best and worst product performance**  
- **Underperforming product categories**  
- **Average customer satisfaction that does not drive stronger sales**  
- **Low shipping costs that do not translate into competitive advantage**

**Tienda 4** stands out as the **least profitable and least strategically valuable store**.

###  **Recommendation: Tienda 4 should be considered for divestment or restructuring**, allowing resources to be reallocated to stronger-performing locations (especially Tienda 1, 2, and 3).
